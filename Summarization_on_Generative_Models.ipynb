```python
from transformers import AutoTokenizer,AutoModelForSeq2SeqLM
!pip install streamlit
import streamlit as st
```

```python
llm="facebook/bart-large-cnn"
tokenizer=AutoTokenizer.from_pretrained(llm)
model=AutoModelForSeq2SeqLM.from_pretrained(llm)
```

```python
st.title("Summarization Application")
input=st.text_area("Enter the paragraph :",height=200)
```

```python
def model_gen1(input):
  tokening=tokenizer(input,return_tensors="pt",truncation=True).input_ids
  output=model.generate(tokening,max_length=100,min_length=50,num_beams=5,early_stopping=True,length_penalty=2.0)
  generated_txt=tokenizer.decode(output[0],skip_special_tokens=True)
  return generated_txt
```

```python
if input:
   summary=model_gen1(input)
   st.subheader("Summary")
   st.write(summary)
```
